```python
import warnings
warnings.filterwarnings('ignore')
from IPython.core.display import display, HTML, Markdown
display(HTML("<style>.container { width:80% !important; }</style>"))
```

```python
import numpy as np
import pandas as pd
import seaborn as sns

tmp = range(15)
data = pd.read_csv('retours_texte.tsv', delimiter = '\t', header = 2, usecols = tmp[-14:])
```

```python
data['En quelle année êtes-vous ?'] = data['En quelle année êtes-vous ?'].astype('category')
```

# Retours textuels en fonction des années


```python
from termcolor import colored


labels_text = ['Quels sont les cours qui vous appréciez suivre à distance ?',
       'Selon vous, pour quelles raisons ces cours sont-ils plus faciles/agréables à suivre ?',
       'Quelles sont les bonnes pratiques à généraliser d\'après vos retours précédents ?',
       'À l\'inverse, quels sont les cours difficiles à suivre et pour quelles raisons ?',
       'Avez vous pu en parler avec le.s enseignant.e.s ? Quelle a été leur réaction.s ? ',
       'Si vous souhaitez laisser un commentaire global, n\'hésitez pas : ']

pd.set_option('display.max_colwidth', None)
md = ''
for annee in data['En quelle année êtes-vous ?'].cat.categories :
    md += '## <span style="color:blue">{}</span>'.format(annee)
    md += '\n'
    df = data.loc[data['En quelle année êtes-vous ?'] == annee]
    for i in labels_text :
        md += '#### <span style="color:red">{}</span>'.format(i)
        md += '\n * '
        md += '\n \n * '.join([str(x) for x in df[i].dropna().values ])
        md += '\n'
        md += '\n'
display(Markdown(md))
```
