## Résultats globaux

```python
def perc_hue(plot, feature, hue):
    numCat = data[feature].nunique()
    numHue = data[hue].nunique()
    patch = [p for p in plot.ax.patches]
    
    a = [p.get_height() for p in patch]
    
    for i in range(numCat):
        total = data[feature].value_counts(sort=False).values[i]
        for j in range(numHue):
            percentage = '{:.1f}%'.format(100 * a[(j*numCat + i)]/total)
            x = patch[(j*numCat + i)].get_x() + patch[(j*numCat + i)].get_width() / 2 - 0.08
            y = patch[(j*numCat + i)].get_y() + patch[(j*numCat + i)].get_height() /2
            plot.ax.annotate(percentage, (x, y), size = 12, color='black')

def perc_no_hue(plot, feature):
    total = len(data[i])
    for p in plot.ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width() /2 - 0.08
        y = p.get_y() + p.get_height()/2
        plot.ax.annotate(percentage, (x, y), color='black')
```

```python
for i in lab :
    plot = sns.catplot(x=i, kind="count", data=data, aspect=18/8)
    perc_no_hue(plot,i)
```

### Résultats en distinguant l'année en cours 

```python
for i in lab :
    plot = sns.catplot(x=i, kind="count", hue="En quelle année êtes-vous ?", data=data, aspect=20/8)
    perc_hue(plot, i, "En quelle année êtes-vous ?" )
```

Le graphe étant peu lisible on va réorienter les labels
